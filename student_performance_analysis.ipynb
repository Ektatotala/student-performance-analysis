```python
import pandas as pd

data = {
    'Student': ['A', 'B', 'C', 'D', 'E'],
    'Math': [80, 70, 90, 60, 85],
    'Science': [75, 65, 88, 70, 80],
    'English': [85, 78, 92, 68, 88]
}

df = pd.DataFrame(data)
df
```

```python
df['Total'] = df[['Math','Science','English']].sum(axis=1)
df['Average'] = df[['Math','Science','English']].mean(axis=1)

df
```

```python
top_student = df.loc[df['Average'].idxmax()]
top_student
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(df['Student'], df['Average'])
plt.title("Average Student Performance")
plt.xlabel("Students")
plt.ylabel("Average Marks")
plt.show()
```

```python
df.set_index('Student')[['Math','Science','English']].plot(kind='bar')
plt.title("Subject-wise Performance")
plt.show()
```

# Insights:
# - Student C has the highest overall performance
